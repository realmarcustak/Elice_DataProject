```python
import pandas as pd
import matplotlib.pyplot as plt

# 7월 데이터 읽기
df_07 = pd.read_csv('서울시_강우량_정보_2022년07월.csv', encoding='euc-kr')
df_07 = df_07[['구청명', '10분우량']]
df_07_grouped = df_07.groupby('구청명').mean()

# 8월 데이터 읽기
df_08 = pd.read_csv('서울시_강우량_정보_2022년08월.csv', encoding='euc-kr')
df_08 = df_08[['구청명', '10분우량']]
df_08_grouped = df_08.groupby('구청명').mean()

# 그래프 그리기
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(16,6))

df_07_grouped.plot(kind='bar', ax=ax1)
ax1.set_title('7월 구별 10분 우량')

df_08_grouped.plot(kind='bar', ax=ax2)
ax2.set_title('8월 구별 10분 우량')

plt.show()
```

```python
# 7월 바 차트
df_07 = pd.read_csv('서울시_강우량_정보_2022년07월.csv', encoding='euc-kr')
df_07 = df_07[['구청명', '10분우량']]
df_07_grouped = df_07.groupby('구청명').mean()

# 8월 바 차트
df_08 = pd.read_csv('서울시_강우량_정보_2022년08월.csv', encoding='euc-kr')
df_08 = df_08[['구청명', '10분우량']]
df_08_grouped = df_08.groupby('구청명').mean()
```

```python
# 바 차트
fig, ax = plt.subplots(figsize=(12, 6))

df_07_grouped.plot(kind='bar', ax=ax, color='red', alpha=0.5, label='7월')
df_08_grouped.plot(kind='bar', ax=ax, color='blue', alpha=0.5, label='8월')

ax.set_title('7월 vs 8월 구별 10분 우량')
ax.set_xlabel('구')
ax.set_ylabel('10분 우량')
ax.legend()

# 선 그래프
fig, ax = plt.subplots(figsize=(12, 6))

df_07_grouped.plot(kind='line', ax=ax, color='red', alpha=0.5, label='7월')
df_08_grouped.plot(kind='line', ax=ax, color='blue', alpha=0.5, label='8월')

ax.set_title('7월 vs 8월 구별 10분 우량')
ax.set_xlabel('구')
ax.set_ylabel('10분 우량')
ax.legend()

plt.show()
```

```python


# csv 파일 읽기
df00 = pd.read_csv('서울시_강우량_정보_2022년07월.csv', encoding='euc-kr')
df = pd.read_csv('서울시_강우량_정보_2022년08월.csv', encoding='euc-kr')

# 필요한 컬럼 추출
df00 = df00[['구청명', '10분우량']]
df = df[['구청명', '10분우량']]

# 구 별로 그룹화하여 평균값 계산
df00_grouped = df00.groupby('구청명').mean()
df_grouped = df.groupby('구청명').mean()

# subplot 생성
fig, ax = plt.subplots(figsize=(12, 6))

# # 바 차트 그리기
# df00_grouped.plot(kind='bar', ax=ax, alpha=0.7, color='r', label='July')
# df_grouped.plot(kind='bar', ax=ax, alpha=0.7, color='b', label='August')

# 선 그래프 그리기
df00_grouped.plot(kind='line', ax=ax, linestyle='--', marker='o', color='r')
df_grouped.plot(kind='line', ax=ax, linestyle='--', marker='o', color='b')

# 제목과 축 라벨 설정
ax.set_title('7월 & 8월 강우량')
ax.set_xlabel('구별')
ax.set_ylabel('강우량')

# 범례 추가
ax.legend()

# 그래프 출력
plt.show()
```

```python


# 7월 데이터 읽기
df_07 = pd.read_csv('서울시_강우량_정보_2022년07월.csv', encoding='euc-kr')
df_07 = df_07[['구청명', '10분우량']]
df_07_grouped = df_07.groupby('구청명').mean()
df_07_grouped = df_07_grouped.rename(columns={'10분우량': '10분우량_7월'})

# 8월 데이터 읽기
df_08 = pd.read_csv('서울시_강우량_정보_2022년08월.csv', encoding='euc-kr')
df_08 = df_08[['구청명', '10분우량']]
df_08_grouped = df_08.groupby('구청명').mean()
df_08_grouped = df_08_grouped.rename(columns={'10분우량': '10분우량_8월'})

# 7월과 8월 데이터 병합
df_merged = pd.merge(df_07_grouped, df_08_grouped, on='구청명')

# 그래프 그리기
fig, ax = plt.subplots(figsize=(12, 6))



df_merged.plot(kind='bar', ax=ax,color=['skyblue', 'darkblue'])
ax.set_title('7월 vs. 8월 구별 10분 우량')
ax.set_ylabel('10분 우량')

plt.show()
```

```python

# 7월 데이터 읽기
df_07 = pd.read_csv('서울시_강우량_정보_2022년07월.csv', encoding='euc-kr')
df_07 = df_07[['구청명', '10분우량']]
df_07_grouped = df_07.groupby('구청명').mean() * 6  # 10분 강우량평균을 1시간 강우량으로 계산

# 8월 데이터 읽기
df_08 = pd.read_csv('서울시_강우량_정보_2022년08월.csv', encoding='euc-kr')
df_08 = df_08[['구청명', '10분우량']]
df_08_grouped = df_08.groupby('구청명').mean() * 6  # 10분 강우량평균을 1시간 강우량으로 계산

# 그래프 그리기
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(16,6))

df_07_grouped.plot(kind='bar', ax=ax1)
ax1.set_title('7월 구별 1시간 강우량')
ax1.set_ylim([0, 1.5]) # y축 범위 설정

df_08_grouped.plot(kind='bar', ax=ax2)
ax2.set_title('8월 구별 1시간 강우량')
ax2.set_ylim([0, 1.5]) # y축 범위 설정

plt.show()
```
